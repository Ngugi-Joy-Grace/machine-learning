## Importing packages and data

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import seaborn as sns
```

```python
gdp=pd.read_csv("Unit04 Global_GDP.csv")
population=pd.read_csv("Unit04 Global_Population.csv")
```

```python
gdp.head(10)
```

```python
population.head(10)
```

```python
# Years range to 2001-2021
years_range = [str(year) for year in range(2001, 2021)]
```

```python
# Converting relevant columns to numeric data type for interpolation
for year in years_range:
    gdp[year] = pd.to_numeric(gdp[year], errors='coerce')
    population[year] = pd.to_numeric(population[year], errors='coerce')
    
# Apply linear interpolation to the relevant columns (2001-2020) for both datasets
gdp[years_range] = gdp[years_range].interpolate(method='linear', axis=1, limit_direction='both')
population[years_range] = population[years_range].interpolate(method='linear', axis=1, limit_direction='both')
```

```python
# Compute the mean GDP and population for each country from 2001 to 2020
mean_gdp = gdp[years_range].mean(axis=1)
mean_population = population[years_range].mean(axis=1)

# Add these computed means to the datasets
gdp['Mean_GDP_2001_2021'] = mean_gdp
population['Mean_Population_2001_2021'] = mean_population
```

```python
gdp[['Country Name', 'Mean_GDP_2001_2021']].head(10)
```

```python
population[['Country Name', 'Mean_Population_2001_2021']].head(10)
```

```python
# Merge the datasets on the "Country Name" column
merged_data = pd.merge(gdp[['Country Name', 'Mean_GDP_2001_2021']],
                       population[['Country Name', 'Mean_Population_2001_2021']],
                       on='Country Name')
```

```python
# Compute the mean per capita GDP for each country from 2001 to 2020
merged_data['Mean_Per_Capita_GDP_2001_2021'] = merged_data['Mean_GDP_2001_2021'] / merged_data['Mean_Population_2001_2021']
```

```python
# Display the merged dataset with the new column
merged_data[['Country Name', 'Mean_GDP_2001_2021', 'Mean_Population_2001_2021', 'Mean_Per_Capita_GDP_2001_2021']].head(10)
```

```python
# Check for duplicated "Country Name" entries in both datasets
gdp_copy = gdp[gdp['Country Name'].duplicated(keep=False)]
population_copy = population[population['Country Name'].duplicated(keep=False)]
```

```python
gdp_copy[['Country Name']]
```

```python
population_copy[['Country Name']]
```

```python
# Remove duplicate entries for Aruba from the population_data dataset
population_cleaned = population.drop_duplicates(subset='Country Name', keep='first')

# Drop rows with NaN as the "Country Name"
population_cleaned = population_cleaned.dropna(subset=['Country Name'])

# Merge the datasets again on the "Country Name" column
merged_cleaned = pd.merge(gdp[['Country Name', 'Mean_GDP_2001_2021']],
                               population_cleaned[['Country Name', 'Mean_Population_2001_2021']],
                               on='Country Name')

# Compute the mean per capita GDP for each country from 2001 to 2020
merged_cleaned['Mean_Per_Capita_GDP_2001_2021'] = merged_cleaned['Mean_GDP_2001_2021'] / merged_cleaned['Mean_Population_2001_2021']

# Display the cleaned merged dataset with the new column
merged_cleaned[['Country Name', 'Mean_GDP_2001_2021', 'Mean_Population_2001_2021', 'Mean_Per_Capita_GDP_2001_2021']].head(10)
```

```python
merged_cleaned
```

```python
# Scatter plot of Mean Population vs Mean Per Capita GDP
plt.figure(figsize=(12, 7))
plt.scatter(merged_cleaned['Mean_Population_2001_2021'], 
            merged_cleaned['Mean_Per_Capita_GDP_2001_2021'], 
            color='blue', 
            alpha=0.6)

plt.title('Mean Population vs. Mean Per Capita GDP (2001-2021)')
plt.xlabel('Mean Population (2001-2021)')
plt.ylabel('Mean Per Capita GDP (2001-2021)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Compute the Pearson Correlation Coefficient between Mean Population and Mean Per Capita GDP
correlation_coefficient = merged_cleaned['Mean_Population_2001_2021'].corr(merged_cleaned['Mean_Per_Capita_GDP_2001_2021'])

correlation_coefficient
```

## Linear Regression


```python
# Drop rows with NaN values
merged_cleaned.dropna(inplace=True)

# Prepare the data
X = merged_cleaned['Mean_Population_2001_2021'].values.reshape(-1, 1)
y = merged_cleaned['Mean_GDP_2001_2021'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Extract model parameters
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)

slope, intercept, r_squared
```
